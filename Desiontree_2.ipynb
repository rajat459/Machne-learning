```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/supply_chain.csv")
```

```python
df.head(1)
```

```python
df.columns
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
categorical_columns = ['Product type', 'SKU', 'Availability','Supplier name','Shipping carriers','Location','Customer demographics','Inspection results','Transportation modes','Routes']

# Applying Label Encoding one by one
label_encoders = {}  # Store encoders for inverse transformation if needed

for col in categorical_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

```python
df.head(1)
```

```python
x=df.drop(columns=['Costs'])
y=df['Costs']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor()
```

```python
dt.fit(x_train,y_train)
```

```python
y_pred=dt.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
param_grid = {
    'max_depth': [10, 20, 30, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
dtree_reg = DecisionTreeRegressor(random_state=42) # Initialize a decision tree regressor
grid_search = GridSearchCV(estimator=dtree_reg, param_grid=param_grid,
                           cv=5, n_jobs=-1, verbose=2, scoring='neg_mean_squared_error')
grid_search.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
best_dtree_reg = grid_search.best_estimator_ # Get the best estimator from the grid search
y_pred = best_dtree_reg.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
rmse = mse ** 0.5
best_params = grid_search.best_params_
print(f"Best parameters: {best_params}")
print(f"Test RMSE: {rmse}")
```
